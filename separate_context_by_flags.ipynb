```python
import pandas as pd
import os
import json
```

```python
# Specify the file path
file_path = "./models/model7/dataset.json"

# Load the JSON data from the file into a Python dictionary
with open(file_path, "r") as json_file:
    list = json.load(json_file)
```

```python
df = pd.DataFrame(list)

def g(df):
    # df['sentiment'] = df['manual_flag'].str.split(',', expand=True)[0]
    # df['type'] = df['manual_flag'].str.split(',', expand=True)[1]
    # df['topic'] = df['manual_flag'].str.split(',', expand=True)[2]
    # df = df.drop(columns=['manual_flag'])
    return df

df = g(df.copy())
df
```

```python
ds = {}

def assign(row):
    # print(row.manual_flag.strip())
    # print(row.comment_display)
    ds.setdefault(row.manual_flag.strip(), [])
    ds[row.manual_flag.strip()].append(row.comment_display)
    
    # ds.setdefault('sentiment', {})
    # ds['sentiment'].setdefault(row.sentiment.strip(), [])

    # ds.setdefault('type', {})
    # ds['type'].setdefault(row.type.strip(), [])

    # ds.setdefault('topic', {})
    # ds['topic'].setdefault(row.topic.strip(), [])

    # ds['sentiment'][row.sentiment.strip()].append(row.comment_display)
    # ds['type'][row.type.strip()].append(row.comment_display)
    # ds['topic'][row.topic.strip()].append(row.comment_display)

df.apply(assign, axis=1)
```

# Create Dataset Directory

```python
dir = './dataset/'
if os.path.exists(dir) == False: os.mkdir(dir)

for label in ds.keys():
    if os.path.exists(dir+'/'+label) == False: os.mkdir(dir+'/'+label)
    print(label+' '+str(len(ds[label])))
    count = 0
    for text in ds[label]:
        f = open(dir+'/'+label+'/'+str(count)+'.txt', "w")
        f.write(text)
        f.close()
        count+=1
```

```python
dir = './dataset_non_null/'
if os.path.exists(dir) == False: os.mkdir(dir)

for label in ds.keys():
    if label != 'null':
        if os.path.exists(dir+'/'+label) == False: os.mkdir(dir+'/'+label)
        print(label+' '+str(len(ds[label])))
        count = 0
        for text in ds[label]:
            f = open(dir+'/'+label+'/'+str(count)+'.txt', "w")
            f.write(text)
            f.close()
            count+=1
```
